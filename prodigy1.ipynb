**I have created my own sample data set in excel file**

```python
import pandas as pd

gender_data = {
    'Gender': ['Male', 'Female', 'Non-Binary', 'Other'],
    'Count': [450, 520, 30, 10]
}
df_gender = pd.DataFrame(gender_data)


age_data = {'Age': [22, 35, 21, 36, 55, 42, 29, 25, 30, 60, 75, 50, 45, 18, 34, 29, 24, 44, 67, 71]}
df_age = pd.DataFrame(age_data)


with pd.ExcelWriter('D:\\population_data.xlsx') as writer:
    df_gender.to_excel(writer, sheet_name='Gender Distribution', index=False)
    df_age.to_excel(writer, sheet_name='Age Distribution', index=False)
```

**Importing the file**

```python
import pandas as pd

file_path = 'D:\\population_data.xlsx'
xls = pd.ExcelFile(file_path)

df_gender = pd.read_excel(xls, 'Gender Distribution')
df_age = pd.read_excel(xls, 'Age Distribution')


print(df_gender.head())
print(df_age.head())
```

**Bar chart for gender**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.bar(df_gender['Gender'], df_gender['Count'], color=['blue', 'pink', 'purple', 'green'])

plt.title('Gender Distribution in Population')
plt.xlabel('Gender')
plt.ylabel('Number of Individuals')


plt.show()
```

**Convert the values into list to plot**

```python
xls = pd.ExcelFile(file_path)

df_age = pd.read_excel(xls, 'Age Distribution')

ages = df_age['Age'].tolist()
```

**Histogram for Age**

```python
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=10, color='skyblue', edgecolor='black')

plt.title('Age Distribution in Population')
plt.xlabel('Age')
plt.ylabel('Number of Individuals')


plt.show()
```
